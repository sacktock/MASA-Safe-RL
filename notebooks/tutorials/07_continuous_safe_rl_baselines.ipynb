## Continuous Cartpole scaffold

`cont_cartpole` is the natural tiny continuous-control environment for this tutorial family. The wrapper stack is still MASA's usual `make_env(...)` path, so labels and constraint metrics live in `info` just like the earlier tutorials.

```python
from masa.plugins.helpers import load_plugins
from masa.common.utils import make_env
from masa.envs.continuous.cartpole import cost_fn, label_fn
from masa.algorithms.ppo import PPO

load_plugins()


def make_continuous_safe_env(max_episode_steps=200):
    return make_env(
        "cont_cartpole",
        "cmdp",
        max_episode_steps,
        label_fn=label_fn,
        cost_fn=cost_fn,
        budget=0.0,
    )


env = make_continuous_safe_env(max_episode_steps=25)
obs, info = env.reset(seed=0)
print("observation_space:", env.observation_space)
print("action_space:", env.action_space)
print("initial observation shape:", getattr(obs, "shape", None))
print("initial labels:", sorted(info["labels"]))
print("initial constraint info:", info["constraint"])
assert env.action_space.shape == (1,)
assert "constraint" in info
assert PPO is not None
env.close()
```
